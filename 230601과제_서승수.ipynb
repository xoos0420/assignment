```python
# 지니 차트 200위까지 크롤링
# 반복문을 사용하여 크롤링
# 수집한 데이터를 데이터프레임으로 저장
# 엑셀로 출력
# 과제는 github로 내보내기
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}

title1 = []
artist1 = []

for i in range(4):
  request = requests.get(f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=17&rtm=Y&pg={i+1}', headers = header)
  soup = BeautifulSoup(request.text)
  bum = soup.find('tbody')
  titles = bum.findAll('a', { 'class' : 'title'})
  artists = bum.findAll('a', { 'class' : 'artist'})
  title1.append(titles)
  artist1.append(artists)
  title2 = sum(title1, [])
  artist2 = sum(artist1, [])

data_dict = {}

def print_rank(titles, artists):
  for j, (t, a) in enumerate(zip(titles, artists)):
    title = t.text.strip()
    artist = a.text.strip()
    data_dict[j] = {'artist' : artist, 'title' : title}

print_rank(title2, artist2)

data_list = list(data_dict.values())
print(data_list)
```

```python
df = pd.DataFrame(data_list)
df
```
